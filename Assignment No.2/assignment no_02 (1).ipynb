```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Amit", "Neha", "Rahul", "Sneha", "Rohit", 
             "Anita", "Vijay", "Neha", "Kiran", "Rahul"],
    "Age": [25, 30, 28, 22, 35, 29, 40, 30, 26, 28],
    "City": ["Pune", "Mumbai", "Delhi", "Pune", "Delhi", 
             "Chennai", "Mumbai", "Mumbai", "Pune", "Delhi"],
    "Salary": [50000, 60000, np.nan, 45000, 70000, 
               52000, 80000, 60000, np.nan, 70000]
}
```

```python
df = pd.DataFrame(data)
print("Original DataFrame:\n", df, "\n")
```

```python
print("DataFrame Info:")
print(df.info(), "\n")
```

```python
print("DataFrame Description:\n", df.describe(), "\n")
```

```python
print("Selected columns (Name, City):\n", df[["Name", "City"]], "\n")
```

```python
df_dropped = df.drop("City", axis=1)
print("After dropping 'City' column:\n", df_dropped, "\n")
```

```python
salary_mean = df["Salary"].mean()
df["Salary"].fillna(salary_mean, inplace=True)
print("After filling missing Salary with mean:\n", df, "\n")
```

```python
df_no_duplicates = df.drop_duplicates()
print("After removing duplicates:\n", df_no_duplicates)
```
