```python
features = df[['genre','rating','members']]
features.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf =TfidfVectorizer (stop_words = 'english')

genre_matrix = tfidf.fit_transform(df['genre'])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

numeric_features = scaler.fit_transform(
    df[['rating','members']]
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(genre_matrix)
```

```python
indices = pd.Series(
    df.index,
    index = df['name']
).drop_duplicates()
```

```python
def recommend_anime(title, top_n=10):

    idx = indices[title]

    sim_scores = list(
        enumerate(cosine_sim[idx])
    )

    sim_scores = sorted(
        sim_scores,
        key=lambda x: x[1],
        reverse=True
    )

    sim_scores = sim_scores[1:top_n+1]

    anime_indices = [i[0] for i in sim_scores]

    return df['name'].iloc[anime_indices]
```

```python
recommend_anime("Naruto")
```

```python
def recommend_threshold(
    title,
    threshold=0.30
):

    idx = indices[title]

    sim_scores = list(
        enumerate(cosine_sim[idx])
    )

    anime_indices = [
        i[0]
        for i in sim_scores
        if i[1] > threshold
    ]

    return df['name'].iloc[anime_indices]
```

Performance Analysis:

Strengths:                                                                                                                                             
• Simple and easy to implement.                                                                                                                        
• Fast recommendation generation.                                                                                                                      
• Uses content information effectively.

Limitations:                                                                                                                                           
• Recommendations depend heavily on genre similarity.                                                                                                  
• User preferences are not directly considered.                                                                                                        
• New anime with limited metadata may not receive accurate recommendations.

Conclusion:

A recommendation system was successfully implemented using cosine similarity.

TF-IDF vectorization converted anime genres into numerical representations. Cosine similarity was used to measure similarity between anime titles and generate recommendations.

The system effectively recommends anime with similar content and demonstrates the usefulness of content-based recommendation techniques.
