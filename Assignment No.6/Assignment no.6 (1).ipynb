```python
import numpy as np
import scipy.stats as stats
```

```python
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
n = len(data)
mean = np.mean(data)
std_sample = np.std(data, ddof=1)   
mean, std_sample, n
```

```python
alpha = 0.01
df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)
margin_error_t = t_critical * (std_sample / np.sqrt(n))
ci_t = (mean - margin_error_t, mean + margin_error_t)
ci_t
```

```python
sigma = 0.2   
z_critical = stats.norm.ppf(1 - alpha/2)
margin_error_z = z_critical * (sigma / np.sqrt(n))
ci_z = (mean - margin_error_z, mean + margin_error_z)
ci_z
```
